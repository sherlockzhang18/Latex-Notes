```python
import numpy as np

def givens_cs(a, b):
    if b == 0.0:
        return 1.0, 0.0
    r = np.hypot(a, b)
    c = a / r
    s = b / r
    return c, s

def qr_rotation(A_init):
    A = A_init.astype(float).copy()
    m, n = A.shape
    Q = np.eye(m)
    for j in range(n):
        for i in range(m-1, j, -1):
            a = A[i-1, j]
            b = A[i, j]
            if abs(b) < 1e-15:
                continue
            c, s = givens_cs(a, b)
            G = np.eye(m)
            G[i-1, i-1] = c
            G[i-1, i]   = s
            G[i,   i-1] = -s
            G[i,   i]   = c
            A = G @ A
            Q = Q @ G.T
    R = A
    return Q, R

A = np.array([
    [8, 5, 1, 0],
    [5, 8, 4, 4],
    [1, 4, 2, 7],
    [0, 4, 7, 4]
], dtype=float)

Q, R = qr_rotation(A)

print("Q:")
print(Q)
print("\nR:")
print(R)
print("\Q R")
print(Q @ R)
```
