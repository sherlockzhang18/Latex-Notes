### Problem 2:

```python
import numpy as np
import matplotlib.pyplot as plt

# building the matix
def build_A(m = 98, n = 5):
    i = np.arange(1, m + 1)[:, None]
    j = np.arange(1, n + 1)[None, :]
    A = np.sin(np.pi * (i - 0.5) * j / m)
    return A.astype(float)

# QR factorization
def qr_factorization(A):
    m, n = A.shape
    Q = np.zeros((m, n))
    R = np.zeros((n, n))

    for j in range(n):
        v = A[:, j].copy() # getting columns of A as v
        
        # removing the components of previous directions
        for i in range(j):
            R[i, j] = np.dot(Q[:, i], v)
            v -= R[i, j] * Q[:, i]
            
        # normalize
        R[j, j] = np.linalg.norm(v, 2)
    return Q, R

m, n = 98, 5
A = build_A(m, n)
Q, R = qr_factorization(A)

# print out results
np.set_printoptions(formatter={'float': '{:0.5f}'.format})
print("R =\n", R)
```

### Problem 3(b):

```python
lam = 2.5 # lambda inside (a)'s calcuated range

def f(x):
    return np.pi - 20.0 * np.arctan(1.0 / x) - 8.0 * np.arctan(3.0 / 79.0)

def relax(x0, lam, max_iter_num = 6):
    res = []
    x = x0
    for k in range(max_iter_num):
        x = x - lam * f(x)
        res.append(x)
    return res

x0 = 9.0
res = relax(x0, lam)

print("Relaxation method:")
for k, x in enumerate(res, start = 1):
    print(f"x{k} = {x:.5f}")
```

### Problem 3(c):

```python
def f(x):
    return np.pi - 20.0 * np.arctan(1.0 / x) - 8.0 * np.arctan(3.0 / 79.0)

def f_prime(x):
    return 20.0 / (x * x + 1.0)

def newton(x0, max_iter_num = 6):
    res = []
    x = x0
    for _ in range(max_iter_num):
        x = x - f(x) / f_prime(x)
        res.append(x)
    return res

x0 = 5.0
res = newton(x0)

print("Newton's Method:")
for k, x in enumerate(res, start = 1):
    print(f"x{k} = {x:.5f}")
```

### Problem 5(a):

```python
import numpy as np

def lu(A: np.ndarray):
    """
    I am doing the LU decomposition without permutation matrix. I am doing it in the order of calculating the j-th row of U first, then the j-th column of L. The values inside the entries are calculated based on the formula provided in the question description. L is initialized to identity matrix and U is initialized to zero matrix.
    """
    A = np.array(A)
    n = A.shape[0]

    # initialize L to be identity matrix (unit lower triangular matrix) and U to be zero matrix
    L = np.eye(n)
    U = np.zeros((n, n))

    for j in range(n):
        # calculating j-th row of U, where j <= q
        # Here I am using row index as j and column index as q for U
        for q in range(j, n): 
            temp = 0.0
            
            # calculate the sum temp variable by the fomula of u_{ij} (the second relation provided in the question)
            for k in range(j):
                temp += L[j, k] * U[k, q]
            U[j, q] = A[j, q] - temp

        # Check if U_{j,j} is zero or not to prevent dividing by zero.
        if U[j, j] == 0.0:
            print("Zero u_jj detected during LU decomposition.")

        # calculating j-th column of L, where i > j
        # Here I am using column index as i and column index as j for L
        for i in range(j + 1, n):
            temp = 0.0
            
            # calculate the sum temp variable by the fomula of l_{ij} (the first relation provided in the question)
            for k in range(j):
                temp += L[i, k] * U[k, j]
            L[i, j] = (A[i, j] - temp) / U[j, j]

    return L, U
```
